```python
## Podglad zebranych danych z GUI (collected_data)
import numpy as np
import matplotlib.pyplot as plt

collected_path = "collected_data/collected_dataset.npz"

data = np.load(collected_path)
x_collected, y_collected = data["images"], data["labels"]

print(f"Liczba zebranych przykladow: {len(y_collected)}")
print(f"Rozklad etykiet: {dict(zip(*np.unique(y_collected, return_counts=True)))}")
n_cols = 8
n_samples = len(y_collected)
n_rows = int(np.ceil(n_samples / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 1.4, n_rows * 1.6))
axes = np.array(axes).reshape(-1)  # spłaszczenie na wypadek n_rows == 1

for i in range(len(axes)):
    ax = axes[i]
    if i < n_samples:
        ax.imshow(x_collected[i].reshape(28, 28), cmap="gray")
        ax.set_title(f"label: {y_collected[i]}", fontsize=9)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
